```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
# lets create synthetic data points
from sklearn.datasets import make_classification
x, y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=2, n_redundant=0)

x
```

```python
y
```

```python
pd.DataFrame(x)[0]
```

```python
sns.scatterplot(
    x=pd.DataFrame(x)[0],
    y=pd.DataFrame(x)[1],
    hue=y
)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC
svc = SVC(kernel='linear')
svc.fit(x_train,y_train)
```

```python
svc.coef_
y_pred = svc.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
s = SVC(kernel='rbf')
s.fit(x_train,y_train)
y_pred1 = s.predict(x_test)
print(classification_report(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
```

```python
polynomial=SVC(kernel='poly')
polynomial.fit(x_train,y_train)
y_pred2=polynomial.predict(x_test)
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
```

```python
sigmoid=SVC(kernel='sigmoid')
sigmoid.fit(x_train,y_train)
y_pred_s=sigmoid.predict(x_test)
print(classification_report(y_test,y_pred_s))
print(confusion_matrix(y_test,y_pred_s))
```

Hyper Parameter Tuning with SVC

```python
from sklearn.model_selection import GridSearchCV

# defining parameters range
param_grid = {
    'C' : [0.1, 1, 10, 100, 1000],
    'gamma' : [1, 0.1, 0.01, 0.001, 0.0001],
    'kernel' : ['rbf']
}

grid = GridSearchCV(SVC(), param_grid = param_grid, refit = True, cv = 5, verbose = 3)

grid.fit(x_train, y_train)
```

```python
grid.best_params_
```

C = regularization parameter

gamma = influence of a single data point

Total Combination : 5(C) * 5(gamma) * 1(kernel) = 25 models

Here we used GridSearchCV to perform hyper parameter tuning

It check all combos of given parameters using cv and selects the model with best performance ......
